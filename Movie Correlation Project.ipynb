```python
# First let's import the packages we will use in this project
# You can do this all now or as you need them
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

# Read in the data
# Replace the file path with the correct path to the 'movies.csv' file on your system
df = pd.read_csv(r'C:\Users\LENOVO\Documents\python\movies.csv')

# Lets look at the data

df.head()



```

```python
# Data cleaning
# To check if there is any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))


```

```python
# Data types for our coloums

df.dtypes


```

```python
#Removing rows with missing values

df=df.dropna()

```

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# change data types of coumns

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')

df['votes'] = df['votes'].astype('int64')

df
```

```python
#Create correct year column

df['released'] = df['released'].str.split('(').str[0]
# .str.split('('): This applies the split() method to each element in the 'released' column, 
#splitting the string at each '(' character. This creates a list of strings.
#.str[0]: This selects the first element (index 0) from the list of strings obtained from the split operation.
#It effectively removes any content after the first '(' character, including the parentheses themselves

df['CorrectYear'] = df['released'].astype(str).str[-5:]

df
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# to display the maximum rows in the table

pd.set_option('display.max_rows', None)
```

```python
# Drop any duplicate

df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
# To check company high correlation 
# budget high correlation


# scatterplot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show()

```

```python
#looking into possible correlations
#Determining correlation between all numeric columns
#Pearson correlation

# Drop non-numeric columns or handle non-numeric values in the DataFrame before calling df.corr()
# For example, you can use df.select_dtypes(include='number').corr() to only include numeric columns for correlation calculation

df.select_dtypes(include='number').corr(method='pearson') #pearson, kendall,spearman
```

```python
# High correlation between budget and gross

correlation_matrix = df.select_dtypes(include='number').corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
#Gross Revenue and other fields

#Converting strings to numbers for correlation calculations

df_numerized = df.copy()
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype=='object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()

```

```python
# correlation for the Numerized Table

# Assuming df_numerized is a DataFrame
correlation_matrix = df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized.corr()
```

```python
# To Unstack

correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs

```

```python
# Alternative to Unstacking is sorting
# To help visualize the correlation of each variable to other variables in the matrix

sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python

# To visualize correlation > 0.5

high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
